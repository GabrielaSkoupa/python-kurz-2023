```python
import pandas
```

```python
zam_praha = pandas.read_csv("zam_praha.csv")
zam_praha["mesto"]= "Praha"
zam_praha.head()

```

```python
zam_plzen = pandas.read_csv("zam_plzen.csv")
zam_plzen["mesto"]= "Plzeň"
zam_plzen.head()
```

```python
zam_liberec = pandas.read_csv("zam_liberec.csv")
zam_liberec["mesto"]= "Liberec"
zam_liberec.head()
```

```python
zamestnanci = pandas.concat([zam_praha, zam_plzen, zam_liberec])
zamestnanci.head()
```

```python
platy_2021_02 = pandas.read_csv("platy_2021_02.csv")
platy_2021_02.head()
```

```python
zam_platy_2021_02 = pandas.merge(zamestnanci, platy_2021_02, on=["cislo_zamestnance"])
zam_platy_2021_02.head()
```
